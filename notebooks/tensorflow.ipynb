```python
import tensorflow as tf
import pandas as pd
import os
from tensorflow import keras
import numpy as np
from datetime import datetime
```

```python
x_train = pd.read_csv('../data/clean/train.csv')
y_train = x_train.SalePrice.values
x_train = x_train.drop('SalePrice', axis=1)
x_train.shape, y_train.shape
```

```python
x_test = pd.read_csv('../data/clean/test.csv').values
```

```python

print(tf.version.VERSION)
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

train_labels = train_labels[:1000]
test_labels = test_labels[:1000]

train_images = train_images[:1000].reshape(-1, 28 * 28) / 255.0
test_images = test_images[:1000].reshape(-1, 28 * 28) / 255.0
train_labels.shape, train_images.shape
```

```python
# Define a simple sequential model

model = tf.keras.models.Sequential([
  keras.layers.Dense(1024, activation='relu', input_shape=(220,)),
  keras.layers.Dense(1024, activation='relu', input_shape=(1024,)),
  keras.layers.Dropout(0.2),
  keras.layers.Dense(1024, activation='relu', input_shape=(1024,)),
  keras.layers.Dropout(0.2),
  keras.layers.Dense(1)
])

model.compile(optimizer='adam',
            loss=tf.losses.mse)
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
price = np.expm1(y_pred) 
price.reshape(-1)
```

```python
submission = pd.DataFrame(np.stack((np.arange(1461, 2920), price.reshape(-1)), axis=1), columns=['Id', 'SalePrice'])
submission['Id'] = submission['Id'].astype(np.int32)
submission.to_csv(f'../submissions/submission_{datetime.today().isoformat()}.csv', index=False)
```
